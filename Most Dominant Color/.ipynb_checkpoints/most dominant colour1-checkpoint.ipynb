```python
import cv2
```

```python
image = cv2.imread("image 1.jpg")
```

```python
image.shape
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
image = image.reshape(image.shape[0] * image.shape[1], image.shape[2])
```

```python
image.shape
```

```python
from sklearn.cluster import KMeans
```

```python
K = 4
```

```python
clf = KMeans(n_clusters=K)
```

```python
clf.fit(image)
```

```python
clf.labels_
```

```python
label = [0, 0, 0, 0]
```

```python
for x in clf.labels_:
    label[x] += 1
```

```python
label_sum = image.shape[0]
label = [float(x) / label_sum for x in label]
```

```python
label
```

```python
def maxi(sequence):
    """return the minimum element of sequence"""
    high = sequence[0] # need to start with some value
    for i in sequence:
        if i > high:
            high = i
    return high
```

```python
maxi(label)
```

```python
clf.cluster_centers_
```

```python
colors = [[int(value) for value in row] for row in clf.cluster_centers_]
```

```python
colors
```

```python
import webcolors
```

```python
clr = [webcolors.rgb_to_hex(c) for c in colors]
```

```python
clr
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(label, colors=clr, shadow=True) 
plt.axis('equal')
plt.show()
```
