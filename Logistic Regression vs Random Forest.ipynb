```python
import pandas as pd
```

```python
df = pd.read_csv("titanic_data_set.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import Imputer
```

```python
imputer = Imputer()
```

```python
df.iloc[:,[5]] = imputer.fit_transform(df.iloc[:,[5]])
```

```python
df.isnull().sum()
```

```python
df.Embarked.value_counts() #for finding the mode
```

```python
df["Embarked"] = df["Embarked"].fillna('S')
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
df.iloc[:,[4]] = labelencoder.fit_transform(df.iloc[:,[4]])
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=["Pclass", "Embarked"])
```

```python
df.head()
```

```python
df.Parch.value_counts()
```

```python
df["FamilySize"] = df["Parch"] + df["SibSp"] + 1
```

```python
df.head()
```

```python
irrelevant_features = ["PassengerId", "Name", "SibSp", "Parch", "Ticket", "Cabin"]
```

```python
df = df.drop(irrelevant_features, axis = 1)
```

```python
df
```

```python
df.to_csv("titanic_processed_data.csv", index=False)
```
